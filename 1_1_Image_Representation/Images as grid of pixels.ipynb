## Importing resources

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import cv2

#%matplotlib qt
```

## Reading in and display the image


```python
#Read in the image
image = mpimg.imread(r"C:\Users\Ashish\Desktop\waymo_car.jpg")

# Print out the image dimension
print("Image dimension :", image.shape)
```

```python
# change from color to grayscale

gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
# Display the image 

plt.imshow(gray_image,cmap='gray')
```

```python
# print specific grayscale pixel value

y = 150
x = 250

pixel_val = gray_image[y, x]

print(pixel_val)
```
